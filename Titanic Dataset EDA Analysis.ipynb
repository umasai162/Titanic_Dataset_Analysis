##  Outlier Analysis

```python
import numpy as np
a=df.select_dtypes(include=np.number)
```

```python
for i in a:
    q1=a[i].quantile(0.25)
    q2=a[i].quantile(0.50)
    q3=a[i].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
Uw=q3+1.5*IQR
Lw=q1-1.5*IQR
```

```python
outliers=((a[i]>Uw)|(a[i]<Lw))
```

```python
print(outliers.sum())
```

```python
a.loc[(a[i]>Uw)|(a[i]<Lw),i]=q2
```

```python
print(outliers.sum())
```
